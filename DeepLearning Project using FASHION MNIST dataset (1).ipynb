In this project, i will be using the Fashion MNIST Hand-Written Dataset which consists of 70,000 images of 10 different clothing categories. This is a large freely available database of fashion images that is commonly used for training and testing various machine learning systems.

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

The Fashion MNIST is divided into a training set and a testing set. In the training set, there are 60,000 images and in the testing set, there are 10,000 images.
The training set which includes (train_images, train_labels) will be used to train the model. The testing set which includes (test_images, test_labels) will be used to evaluate the model’s performance on new, unseen data.

Our aim is to build an artificial neural network model that will classify the images correctly to their labels. For example, if we feed the model with a “Coat” image it should classify that as the value 4 because “Coat” are given the label 4 in our dataset.

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

We will use the training set which includes (train_images, train_labels) to train our model. Then we will use the testing set which includes (test_images, test_labels) to evaluate our model’s performance on new, unseen data.

```python
train_images.shape, train_labels.shape #((60000, 28, 28), (60000,))
test_images.shape, test_labels.shape #((10000, 28, 28), (10000))
```

```python
plt.figure()
plt.imshow(train_images[0], cmap='binary')
plt.colorbar()
plt.xticks([])
plt.yticks([])
```

I used matplotlib to visualize the image. The images are gray-scale each pixel will have a value between 0–255. 0 means WHITE. 255 means BLACK.

```python
# Data preprocessing
train_images = train_images/255.0
test_images = test_images/255.0
```

The above code will reduce the pixel value from 0–255 to 0–1. Its good to reduce the value of our pixels before proceeding with model creation. Zero means white and 1 means black.

```python
plt.figure()
plt.imshow(train_images[0], cmap='binary')
plt.colorbar()
plt.xticks([])
plt.yticks([])
```

 The color bar on the right shows a different range.

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

Lets print the first 25 images in our dataset

```python

plt.figure(figsize=(10, 10))
for i in range(25):
  plt.subplot(5, 5, i+1)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(train_images[i], cmap='binary')
  plt.xlabel(class_names[train_labels[i]])
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
model.compile(optimizer='adam', loss="sparse_categorical_crossentropy",
              metrics=['accuracy'])
```

We created a neural network with 3 layers. The first layer is the input layer which is all the pixels in our image. The second layer is a hidden layer with 128 neurons(it is just an arbitrary value) this is where the computation takes place (here is where the weights and the biases are found). The third layer is the output layer with 10 neurons(because we have 10 different categories).

When compiling the model, we need 2 important things a loss function and an optimizer. We need a loss function to know how far our prediction is from the original value. We need an optimizer to reduce the distance between the prediction and the original value.

```python
# Training the model
model.fit(train_images, train_labels, epochs=10)
```

```python
model.evaluate(test_images, test_labels)
```

this shows an accuracy of 87% in unseen (test set) data.

```python
# prediction
classification = model.predict(test_images)
```

Now our prediction is stored in the variable called classification.

Let’s see the prediction for the 0th index.

```python
classification[0]
```

See the above output. It says the probability of the image being in category 1(T-shirt/Top) is 2.5. The probability of the image being in category 2(Trousers) is 9.6 and so on.
Category 9 (ankle boot) has the highest probability.
So, the image in the 0th index is an Ankle Boot. Let’s see that in code.

```python
np.argmax(classification[0]), test_labels[0]
```

np.argmax() will give us the index of the highest value in the NumPy array.

So, the prediction is also 9 and the original value is also 9.
